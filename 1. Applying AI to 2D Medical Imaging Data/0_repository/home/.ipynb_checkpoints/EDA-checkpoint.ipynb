### Task 1: Demographic plots

Histogram plots of following attributes are visualized:
    1. Patient Age
    2. Patient Gender
    3. View Position (X-Ray views)

```python
all_xray_df["Patient Age"].hist(bins=20, align="left", rwidth=0.9, figsize=(7, 6))
```

```python
# removing over aged patients
all_xray_df = all_xray_df[all_xray_df['Patient Age'] <= 120]
all_xray_df.sample(3)
```

```python
# create sum of diseases
all_xray_df['disease_sum'] = all_xray_df.iloc[:, 12:26].sum(axis=1)
all_xray_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 7))
axs = axs.ravel()

all_xray_df["Patient Age"].hist(bins=20, align="left", rwidth=0.9, ax=axs[0])
axs[0].set_xlabel("Patient Age")
axs[0].set_ylabel("Frequency")
axs[0].set_title("Histogram of Patient Age Distribution in the dataset")
axs[0].grid(True, alpha=0.5)

all_xray_df["Patient Gender"].hist(bins=range(3), align="left", rwidth=0.7, ax=axs[1])
axs[1].set_xlabel("Patient Gender")
axs[1].set_ylabel("Frequency")
axs[1].set_title("Histogram of Patient Gender Distribution in the dataset")
axs[1].grid(True, alpha=0.5)

df = all_xray_df[(all_xray_df.Pneumonia == 1) & (all_xray_df.disease_sum == 1)]
df["Patient Age"].hist(bins=20, align="left", rwidth=0.9, ax=axs[2])
axs[2].set_xlabel("Patient Age")
axs[2].set_ylabel("Frequency")
axs[2].set_title("Histogram of Patient Age Distribution (Pneumonia only)")
axs[2].grid(True, alpha=0.5)

df["Patient Gender"].hist(bins=range(3), align="left", rwidth=0.7, ax=axs[3])
axs[3].set_xlabel("Patient Gender")
axs[3].set_ylabel("Frequency")
axs[3].set_title("Histogram of Patient Gender Distribution (Pneumonia only)")
axs[3].grid(True, alpha=0.5)

fig.tight_layout()
plt.savefig("Age and Gender distribution in dataset.png")
```

**CONCLUSION**
* Patient Age distribution: Certain patients seem to have age more that 120. Demographic data suggests that age of human
beings can rarely be more than 100 years and only age of 2% people in United States of America is more than 85 years. **Therefore all the patient data above 120 years will be removed.**
* Most patients after removing data are between the age of 40 - 60 years old
* The age distribution for pneumonia-only patient follows a comparatively uniform distribution in the age range 20-60 years.
* Gender distribution suggests that there are more men patients than women patients.
* The gender distribution follows same distribution in patients with only pneumonia

### Task 4: Diseases comorbid with Pneumonia

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 7))
axs = axs.ravel()

df = all_xray_df[all_xray_df['disease_sum'] == 1]
df['Finding Labels'].value_counts().plot(kind="bar", grid=False,
                                         title="Patients having only one disease",
                                         rot=90, ax=axs[0])
axs[0].set_xlabel("Disease type")
axs[0].set_ylabel("Frequency")
axs[0].grid(True, alpha=0.5)

df = all_xray_df[(all_xray_df['disease_sum'] < 3) & (all_xray_df['Pneumonia'] == 1)]
df["Finding Labels"].value_counts().plot(kind="bar", grid=False,
                                         title="Distribution of diseases comorbid with pneumonia",
                                         rot=90, ax=axs[1])
axs[1].set_xlabel("Disease with Pneumonia")
axs[1].set_ylabel("Frequency")
axs[1].grid(True, alpha=0.5)

fig.tight_layout()
plt.savefig("distribution of diseases.png")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 14))
df = all_xray_df[all_xray_df['disease_sum'] != 0]
df.disease_sum.value_counts().plot(kind="bar", grid=False, ax=axs[0][0],
                                   title="Distribution of Number of Diseases in SICK Patients",
                                   rot=0)

df = all_xray_df[all_xray_df['disease_sum'] == 1]
df['Finding Labels'].value_counts().plot(kind="bar", grid=False, ax=axs[0][1],
                                         title="Patients having only one disease",
                                         rot=90)

df = all_xray_df[all_xray_df['Pneumonia'] == 1]
df.disease_sum.value_counts().plot(kind="bar", grid=False, ax=axs[1][0],
                                   title="Number of Disease Distribution in Pneumonia Patients",
                                   rot=90)

df = all_xray_df[(all_xray_df['disease_sum'] < 3) & (all_xray_df['Pneumonia'] == 1)]
df["Finding Labels"].value_counts().plot(kind="bar", grid=False, ax=axs[1][1],
                                         title="Number of Disease Distribution in Pneumonia Patients",
                                         rot=90)

fig.tight_layout()
```

```python
fig, m_axs = plt.subplots(1,2, figsize = (8, 4))
pn_case = pneumonia_cases.iloc[np.random.randint(0, len(pneumonia_cases))]
p_img = Image.open(pn_case.path)
m_axs[0].imshow(p_img, cmap='gray')
m_axs[0].set_title(pn_case['Finding Labels'])

no_pn_case = no_pneumonia_cases.iloc[np.random.randint(0, len(no_pneumonia_cases))]
np_img = Image.open(no_pn_case.path)
m_axs[1].imshow(np_img, cmap='gray')
m_axs[1].set_title(no_pn_case['Finding Labels'])

fig.tight_layout()
```

```python
# For Pneumonia case
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.ravel()

pixel_array = np.array(p_img.getdata())
mean = np.mean(pixel_array)
std = np.std(pixel_array)
axs[0].hist(pixel_array.ravel(), bins=256)
axs[0].set_title('Pixel Intensity Distribution in Pneumonia Patient')
axs[0].text(0, 7500, "mean: {:2f}\nstd: {:2f}".format(mean, std), fontsize=12)

new_img = pixel_array.copy()
normalized_array = (new_img - mean) / std
axs[1].hist(normalized_array.ravel(), bins=256)
axs[1].set_title('Normalized Pixel Intensity Distribution in Pneumonia Patient')

pixel_array = np.array(np_img.getdata())
mean = np.mean(pixel_array)
std = np.std(pixel_array)
axs[2].hist(pixel_array.ravel(), bins=256)
axs[2].set_title('Pixel Intensity Distribution in non-Pneumonia Patient')
axs[2].text(0, 9500, "mean: {:2f}\nstd: {:2f}".format(mean, std), fontsize=12)

new_img = pixel_array.copy()
normalized_array = (new_img - mean) / std
axs[3].hist(normalized_array.ravel(), bins=256)
axs[3].set_title('Normalized Pixel Intensity Distribution in non-Pneumonia Patient')

fig.tight_layout()
plt.savefig("pixel_intensity_distribution.png")
```

### Conclusion
It is difficult to tell from intensity distribution as other diseases interfere with pixel intensity distribution. So it is difficult see from intensity distribution if patient has pneumonia or other disease or non. Moreover, intensity distribution of only-pneumonia patients has two peaks compared to just one peak in healthy patients. So they can be identified comparatively with ease.
